```python
def generate_kmer(qualities, order, readsize):
    starts = []
    kmers = {}
    for q in qualities:
        for i in range(0, len(q)-order):
            kmer = q[i:i+order]
            if i == 0:
                starts.append(kmer)
            if kmer not in kmers.keys():
                kmers[kmer] = []
            kmers[kmer].append(q[i+order])
    return([starts, kmers])


def mchain(starts, kmers, readsize, order):
    seed = random.choice(starts)
    while(seed not in kmers.keys()):
        seed = random.choice(starts)
    chain = seed
    while(len(chain) < readsize):
        if seed in kmers.keys():
            first = kmers[seed]
            nxt = random.choice(first)
            chain += nxt
            seed = chain[len(chain)-order:len(chain)]

        else:
            return(0)
    return(chain)
```

```python
READLEN = 76
MARKOV_ORDER = 10
QUALIT = pickle.load(open('qualities.p', 'rb'))
MARKOV_SEED = generate_kmer(qualities=QUALIT, order=MARKOV_ORDER, readsize=READLEN)
MARKOV_START = MARKOV_SEED[0]
MARKOV_DICT = MARKOV_SEED[1]
```

```python
def markov_wrapper(times, starts, kmers, readsize, order):
    a = 0
    while(a == 0):
        a = mchain(starts=starts, kmers=kmers,
                      readsize=readsize, order=order)
    return(a)


def markov_multi(process, nreads, starts, kmers, readsize, order):
    myIter = range(nreads)
    markov_partial = partial(
        markov_wrapper, starts=starts, kmers=kmers, readsize=readsize, order=order)
    with multiprocessing.Pool(process) as p:
        r = list(p.imap_unordered(markov_partial, myIter))
    return(r)
```
